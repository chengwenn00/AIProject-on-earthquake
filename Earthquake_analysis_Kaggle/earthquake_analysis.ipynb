# Cleaning Predictor

```python
trainData_clean = trainData.drop(columns = trainData.columns[26:])
trainData_clean = trainData_clean.drop(columns = trainData_clean.columns[10])
trainData_clean = trainData_clean.drop(columns = trainData_clean.columns[0])
trainData_clean = trainData_clean.replace({'land_surface_condition': {'n':1, 'o':2, 't':3}})
trainData_clean = trainData_clean.replace({'foundation_type': {'h':1, 'i':2, 'r':3, 'u':4, 'w':5}})
trainData_clean = trainData_clean.replace({'roof_type': {'n':1, 'q':2, 'x':3}})
trainData_clean = trainData_clean.replace({'ground_floor_type': {'f':1, 'm':2, 'v':3, 'x':4, 'z':5}})
trainData_clean = trainData_clean.replace({'other_floor_type': {'j':1, 'q':2, 's':3, 'x':4}})
trainData_clean = trainData_clean.replace({'position': {'j':1, 'o':2, 's':3, 't':4}})
trainData_clean = trainData_clean.replace({'plan_configuration': {'a':1, 'c':2, 'd':3, 'f':4, 'm':5, 'n':6, 'o':7, 'q':8, 's':9, 'u':10}})
trainData_clean.info()
```

# Function For ClassificationTree

```python

def classtree(var):
    # Import train_test_split from sklearn
    from sklearn.model_selection import train_test_split
    # Split the Dataset into Train and Test
    X_train, X_test, y_train, y_test = train_test_split(var, damageData, test_size = 0.2)
    # Import DecisionTreeClassifier model from Scikit-Learn 
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.metrics import confusion_matrix
    
    # Check the sample sizes
    print("Train Set :", y_train.shape, X_train.shape)
    print("Test Set  :", y_test.shape, X_test.shape)
    # importing Random forest classifier from assemble module
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.metrics import f1_score
    import pandas as pd
    # creating a RF classifier
    clf = RandomForestClassifier(n_estimators = 100) 

    # Training the model on the training dataset
    # fit function is used to train the model using the training sets as parameters
    clf.fit(X_train, y_train)
    # performing predictions on the test dataset
    y_pred = clf.predict(X_test)

    # metrics are used to find accuracy or error
    from sklearn import metrics 
    print()
    # Build the random forest on the full training data
    rf = RandomForestClassifier() # instance
    rf.fit(X_train, y_train) # fit model
    y_pred = rf.predict(X_test) # predict on test
    
    # Calculate confusion matrix
    con_mat = confusion_matrix(y_test, y_pred)
    con_mat = con_mat / con_mat.astype(np.float).sum(axis=1)

    # Evaluate model
    print("F1 Score: %f " % f1_score(y_test, y_pred,average='micro'))

    from sklearn.metrics import confusion_matrix #plotting confusion matrix
    sb.heatmap(confusion_matrix(y_test, y_pred),
          annot = True, fmt=".0f", annot_kws={"size": 18})
    # Training the model on the training dataset
    # fit function is used to train the model using the training sets as parameters
    clf.fit(X_train, y_train)
    # performing predictions on the test dataset
    y_pred = clf.predict(X_test)
    # metrics are used to find accuracy or error
    from sklearn import metrics 
    print()
    
    cmtrain = confusion_matrix(y_test, y_pred)
    damage1actual = cmtrain[0][0] + cmtrain[0][1] + cmtrain[0][2]
    damage2actual = cmtrain[1][0] + cmtrain[1][1] + cmtrain[1][2]
    damage3actual = cmtrain[2][0] + cmtrain[2][1] + cmtrain[2][2]
    print('X = ', damage3actual + damage2actual - damage1actual)
```

# Bernoulli Classifier vs Decision Tree Classifier vs Random Forest Classifier

# Decision Tree Classifier

```python
# Import DecisionTreeClassifier model from Scikit-Learn
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
# Import train_test_split from sklearn
from sklearn.model_selection import train_test_split

# Split the Dataset into Train and Test
X_train, X_test, y_train, y_test = train_test_split(trainData_clean, damageData, test_size = 0.2)

# Check the sample sizes
print("Train Set :", y_train.shape, X_train.shape)
print("Test Set  :", y_test.shape, X_test.shape)

# Import DecisionTreeClassifier model from Scikit-Learn
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

# Decision Tree using Train Data
dectree = DecisionTreeClassifier(max_depth = 1000)  # create the decision tree object
dectree.fit(X_train, y_train)                    # train the decision tree model
# Predict Legendary values corresponding to Total
y_train_pred = dectree.predict(X_train)
y_test_pred = dectree.predict(X_test)

# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Classification Accuracy \t:", dectree.score(X_train, y_train))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Classification Accuracy \t:", dectree.score(X_test, y_test))
print()

# Plot the Confusion Matrix for Train and Test
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(confusion_matrix(y_train, y_train_pred),
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(confusion_matrix(y_test, y_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])
```
